### 2. 윌콕슨의 부호순위 검정(Wilcoxon's Signed Rank Test)

```
# 사전의 양적 자료에서 사후의 양적 자료를 뺀 값이
# 정규성 가정을 만족하지 않을 때 사용. 즉 정규성 가정이 깨졌을 때 사용
```

```
# 가설검정 01
# 귀무가설 수업의 효과가 없다
# 대립가설 수업의 효과가 있다
```

```
# stats.wilcoxon(pre, post, zero_method = 'wilcox', alternative = '') # zero_method = 'wilcox' pre - post 가 0인 값은 분석에서 제외
stats.wilcoxon(satisfy.pre, satisfy.post, zero_method='wilcox', alternative='less')
```

```
# 결론

# 검정통계량 = W+ = 0.000 (데이터 양이 작기때문에 t값이 아닌 W+ 값을 검정통계량으로 출력)
# 유의확률(p-value) = 0.008

# 유의확률이 0.008이므로 유의수준 0.05에서 수업의 효과는 통계적으로 유의미한 수준인 것으로 나타났다.
```

```
# 가설검정 02
# 귀무가설 : 다이어트는 몸무게를 줄이는 것에 효과가 없다
# 대립가설 : 다이어트는 몸무게를 줄이는 효과가 있다.
```

```
stats.wilcoxon(diet.before, diet.after, zero_method='wilcox', alternative='greater')
```

```
# 결론
# 유의확률 0.006이므로 유의수준 0.05에서 다이어트 이후 몸무게 감소가 통계적으로 유의미한 수준인 것으로 나타났다.
```

### 실행단계

```
# 만족도(Satisfy)
# 1단계 : 정규성 검정
# 귀무가설 : D ( = pre - post)가 정규분포를 따른다
# 대립가설 : D가 정규포를 따르지 않는다.
stats.shapiro(satisfy.difference)
```

```
# 결론
# 유의확률이 0.195 이므로 유의수준 0.05에서 정규성 가정을 만족한다.
```

```
# 2단계 : Paired t-test
stats.ttest_rel(satisfy.pre, satisfy.post)
```

```
# 결론 : 유의 확률이 0.004이므로 유의수준 0.05에서 수업의 효과는 통계적으로 유의미한 것으로 드러났다.
```

```
# Diet

# 1단계 : 정규성 검정
# 귀무가설 : (before - after)가 정규분포를 따른다
# 대립가설 : 따르지 않는다
stats.shapiro(diet.difference)
```

```
#유의확률이 0.011이므로 유의수준 0.05에서 정규성 가정이 깨짐
```

```
#Wilcoxon 검정 시행
stats.wilcoxon(diet.before, diet.after, zero_method='wilcox', alternative='greater')
```

```
#결론: 유의확률이 0.006이므로 유의수준 0.05에서 다이어트 후의 몸무게는 다이어트 이전에 비해 통계적으로 유의미한 수준으로 감소하였다.
```
