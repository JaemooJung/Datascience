```
import pandas as pd
import scipy.stats as stats
diamonds = pd.read_excel('./diamonds.xlsx', sheet_name = 0, header = 0)
diamonds
```

```
# 귀무가설 : cut과 color 간에는 관련성이 없다!
# 대립가설 : cut과 color 간에는 관련성이 있다!
```

```
# 참고
# cut = 질적자료 : 5개의 범주 (Category)
# color = 질적자료 : 7개의 범주 (Category)
```

```
# result = pandas.crosstab(d1, d2)
# stats.chi2_contingency(observed = result)
```

```
result = pd.crosstab(diamonds.cut, diamonds.color)
```

```
result.head()
```

```
stats.chi2_contingency(observed = result)
```

```
# 결과해석 : 
# 첫 번째 ; 310.31790052115434    = Chi-square
# 두 번째 ; 1.394512091985105e-51 = p-value
# 세 번째 ; 24                    = df(자유도)
# 네 번째 ; Expected values
```

```
# 결론 : 유의확률 0.000으로 유의수준 0.05에서 다이아몬드의 cut과 color 사이에는 통계적으로 유의미한 관련성이 있는 것으로 나타났다. 
```

```
# 귀무가설 : cut과 clarity 사이에는 관련성이 없다
# 대립가설 : 관련성이 있다
```

```
result2 = pd.crosstab(diamonds.cut, diamonds.clarity)
stats.chi2_contingency(result2)
```

```
# 결론
# 유의확률 0.000으로 유의수준 0.05에서 다이아몬드의 cut과 clarity 사이에는 통계적으로 유의미한 관련성이 있는 것으로 나타났다.
```
