## 회귀분석(Regression Analysis)

```
# 양적자료(X) 혹은 양적자료들(Xs)이 다른 양적자료(Y)에게 영향을 주는 지를 분석하는 방법
# 분석방법
# 인과관계(원인과 결과)를 분석하는 방법
# 예측(Prediction, Forecasting)할 때 주로 사용하는 방법
# 분류(Classification)할 때도 사용됨.

# 영향을 주는 양적자료 : X = feature = 독립변수 = 설명번수
# 영향을 받는 양적자료 : Y = label = 종속변수 = 반응변수
```

```
# 기본
# Y : 양적자료
# X : 양적자료
```

```
# 기본
# Y와 X간의 선형으로 접근
```

```
# 회귀분석의 종류
# 1. 단순선형회귀분석(Simple linear regression analysis)
# Y : 1개
# X : 1개 (그러나 실전에서는 X가 하나인 경우는 거의 없음)

# 2. 다중선형회귀분석(Multiple linear regression analysis)
# Y : 1개
# X : 2개 이상
```

### 1. 단순선형 회귀분석(Simple Linear Regression Analysis)

```
# 예제데이터 : iris
iris.head()
```

```
# model = sm.OLS(endog = Y, exog = X).fit()
# model.summary()

# 참고
# OLS = Ordinary Least Square = 회귀분석
# endog = endogenous : 내생적인, 내인성의 = 종속변수
# exog = exogenous : 외생적인, 외인성의 = 독립변수

# Y = petal_length : 꽃잎의 길이
# X = sepal_length : 꽃받침의 길이
```

```
iris_model = sm.OLS(endog= iris.petal_length, exog= iris.sepal_length).fit()
iris_model.summary()
```

#### 회귀분석 결과의 단계적 해석

```
# 1단계 : 회귀모형은 타당한가?
# 귀무가설 : 회귀모형은 타당하지 않다.
# 대립가설 : 회귀모형은 타당하다. 
```

```
# 회귀분석 결과
# F-statistic:	1334.           / -> 검정통계량 F
# Prob (F-statistic):	3.02e-76  / ->  유의확률 = 0.000

# 결론
# 유의확률이 0.000이므로 유의수준 0.05에서 회귀모형은 통계적으로 유의하게 타당한 것으로 나타났다.
```

```
# 2단계 : X는 Y에게 유의한 영향을 주는가?
# 귀무가설 : X는 Y에게 영향을 주지 않는다
# 대립가설 : X는 Y에게 영향을 준다
```

```
# 회귀분석 결과

#               coef	  std err	  t	      P>|t|	   [0.025	0.975]
# sepal_length	0.6669	0.018	    36.530	0.000	    0.631 0.703


# coef    : 회귀계수(beta 1)         = 0.667  (beta1 hat)
# std err : 표준오차(standard error) = 0.018  (다른 표본을 통해 b1 hat을 구했을 때의 표준오차)
# t       : 회귀계수 / 표준오차       = 36.530
# P > |t| : 유의확률                 = 0.000

# 결론
# 유의확률이 0.000이므로 유의수준 0.05에서 X는 Y에 유의한 영향을 주는 것으로 나타났다.
# 유의확률이 0.000이므로 유의수준 0.05에서 꽃받침의 길이는 꽃잎의 길이에 통계적으로 유의한 영향을 주는 것으로 나타났다.
```

```
# 3단계 : X는 Y에게 어떠한 영향을 주는가?
```

```
# 회귀분석 결과
#               coef	  
# sepal_length	0.6669

# coef : coefficient of regression (beta 1) : 회귀계수
# 회귀계수 : 0.667
# 회귀계수의 값이 크면, X가 Y에게 영향을 많이 준다.
# 회귀계수의 값이 작으면 X가 Y에게 미미한 영향을 준다.
# 회귀계수의 부호가 + 이면 X가 Y를 증가시키는 영향을 준다.
# 회귀계수의 부호가 - 이면 X가 Y를 감소시키는 영향을 준다.

# 해석
# X의 기본단위가 1 증가하면 Y는 약 0.667 정도 증가시킨다.

# -> 꽃받침의 길이가 1cm 증가하면 꽃잎의 길이는 약 0.667cm 증가시키는 영향을 준다
```

```
# 4단계 : 회귀모형의 설명력 또는 X의 설명력

# Y의 다름(변동)을 회귀모형이 얼마나 설명하고 있는가?
# = Y의 다름(변동)을 X가 얼마나 설명하고 있는가?
```

```
# 회귀분석 결과
# R-squared (uncentered):	0.900

# R-square : 결정계수(Coefficient of Determination) = 0.900
# R-square = SSR / SST : 0 ~ 1 
# R-square 값이 크면 회귀모형이 좋은 것으로 판단하는 지표 중 하나임

# 해석
# 회귀모형이 Y의 다름(변동)을 약 90.0% 정도 설명하고 있다.
# X가 Y의 다름(변동)을 90.0% 정도 설명하고 있다.
# 꽃받침의 길이가 꽃잎의 길이의 변동을 약 90.0% 정도 설명하고 있다. 

# *참고 -> R-square 값이 충분한가? 는 도메인에 따라 달라짐. 통제된 실험실환경에서 측정된 것이라면 1에 아주 가까운 값이어야 의미가 있겠지만, 사회과학 연구에서 0.9면 1과 다름없음...
```

```
# 5단계 : 예측(Prediction or Forecasting)
# model.predict(X)
iris_predict = iris_model.predict(exog = iris.sepal_length)

print(iris_predict)
print(iris.sepal_length)
```

```
# MSE = Mean Squared Error (???)
# MSE = SSE / df = SSE / n-p-1 = SSE / n-2 
# from sklearn.metrics import mean_squared_error
# mean_square_error(y_true = , y_pred = )
mean_squared_error(y_true= iris.petal_length, y_pred=iris_predict) 
```

```
# 6단계 : 회귀모형의 진단 또는 회귀모형에 대한 가정
# (1) 오차의 독립성 가정
# (2) 오차의 정규성 가정
# (3) 오차의 등분산성 가정
# (4) 오차의 선형성 가정
```

```
# (1) 오차의 독립성 가정

# Durbin-Watson: 0.101 -> 오차들이 독립인지를 알려주는 값.

# DW 값이 0에 가까워지면 오차들은 독립이 아니고, 양의 자기상관이 있다.
# DW 값이 2에 가까우면 오차들은 독립이다.
# DW 값이 4에 가까워지면 오차들은 독립이 아니고, 음의 자기상관이 있다.

# 독립성 검정
# 귀무가설 : 독립이다.
# 대립가설 : 독립이 아니다.
# statsmodels.stats.stattools.durbin_watson(residuals)

# 참고
# Residual(잔차) : 실제 Y - 예측 Y  

sms.stattools.durbin_watson(iris_model.resid)

# 결론
# 독립성 가정은 깨졌다.
```

```
# (2) 정규성 검정
# 귀무가설 : 정규분포를 따른다
# 대립가설 : 정규분포를 따르지 않는다.

# Jarque-Bera (JB):	16.638
# Prob(JB):	0.000244
# P-value = 0.000 이므로 유의수준 0.05에서 정규성 가정이 깨짐

# stats.shapiro(residual)
stats.shapiro(iris_model.resid)

# 결론 : 유의 확률이 0.000이므로 유의수준 0.05에서 정규성 가정이 깨짐.
```

```
# Omnibus :210.387
# Prob(Omnibus):	0.000

# 모든 독립변수들을 회귀모형에 사용했을 때 회귀모형의 유의성을 검정함.
# 귀무가설 : 유의하지 않다.
# 대립가설 : 유의하다.

# 결론
# 유의확률이 0.000
# 따라서 회귀모형은 통계적으로 유의한 것으로 나타났다. 
```

```
# (3) 등분산 검정
# 오차에 대한 등분산검정
# 귀무가설 : 등분산이다.
# 대립가설 : 등분산이 아니다. 

# stats.bartlett(X, residuals)
stats.bartlett(iris.sepal_length, iris_model.resid)

# 유의확률이 0.000이므로 유의수준 0.05에서 등분산 가정이 깨짐.
```

### 2. 다중선형 회귀분석(Multiple Linear Regression Analysis

```
# Y : 양적 자료 1개
# X : 양적 자료 2개 이상
```

```
# 예제 데이터 : iris
# Y : petal_length
# X : 나머지

Y = iris.petal_length
X = iris.loc[:, ['sepal_length','sepal_width','petal_width']]
```

```
iris_mlr_model = sm.OLS(endog= Y,
                        exog = X).fit()
iris_mlr_model.summary()
```

```
#회귀분석결과의 해석
```

```
# 1단계 : 회귀모형의 타당성 검정
# 귀무가설 : 회귀모형은 타당하지 않다
# 대립가설 : 회귀모형은 타당하다

# F-statistic: 8426.
# Prob (F-statistic):	3.14e-164

# 결론 
# 유의확률 0.000으로 유의수준 0.05에서 회귀모형은 통계적으로 유의하게 타당한 것으로 나타났다.
```

```
# 2단계 : X들의 유의성 검정 ; X가 Y에게 영향을 주는가?
# 귀무가설 : X는 Y에게 영향을 주지 않는다 = Beta_i == 0
# 대립가설 : X는 Y에게 영향을 준다 = Beta_i != 0

#	              coef	  std err	  t	      P>|t|	  [0.025	0.975]
#sepal_length	  0.6942	0.043	    16.205	0.000	  0.610	 0.779
#sepal_width	 -0.6729	0.061	   -10.969	0.000	 -0.794	-0.552
#petal_width	  1.4680	0.063	    23.247	0.000	  1.343	 1.593

# 각각의 X에 대해 t검정 실시
# 세 X 모두 pvalue 0.000으로, 모든 X는 유의수준 0.05에서 Y에  통계적으로 유의한 영향을 미치는 것으로 나타났다.
```

```
# 3단계 : X(들)는 Y에게 어떠한 영향을 주는가?

#	              coef	  
#sepal_length	  0.6942	
#sepal_width	 -0.6729	
#petal_width	  1.4680	

# sepal_length : 0.694
# sepal_width : -0.673
# petal_width : 1.468

# 꽃받침의 너비와 꽃잎의 너비가 고정되어 있을 때, 꽃받침의 길이가 1 cm 증가하면 꽃잎의 길이는 약 0.694cm 증가한다.
# 꽃받침의 길이와 꽃잎의 길이가 고정되어 있을 때, 꽃받침의 너비가 1 cm 증가하면 꽃잎의 길이는 약 0.673cm 감소한다.
# 꽃받침의 길이와 꽃잎의 너비가 고정되어 있을 때, 꽃잎의 너비가 1 cm 증가하면 꽃잎의 길이는 약 1.468cm 증가한다.
```

```
# 4단계 : 회귀모형의 설명력

# R-squared (uncentered)     :	0.994
# Adj. R-squared (uncentered):	0.994

# 결정계수는 X가 증가하면 결정계수는 무조건 증가함.

# Adj. R-squared = Adjusted R-squared = 수정된 결정계수
# 수정된 결정계수는 Y에게 유의한 X가 모형에 들어올 때는 결정계수가 증가하고
# Y에세 유의하지않은 X가 모형에 들어올 때는 결정계수가 증가하지 않도록 조치를 한 것임.

# Adj. R-squared (uncentered):	0.994
# 따라서 회귀모형이 Y의 변동을 약 99.4% 가량 설명하고 있다.
# 꽃받침의 길이, 꽃받침의 너비, 꽃잎의 너비가 꽃잎의 길이의 변동을 약 99.4% 가량 설명하고있다.
```

```
# 5단계 : 예측
iris_predict = iris_mlr_model.predict(exog = X)
iris_predict
```

```
# MSE는 작을수록 모형이 좋다고 판단함.
mean_squared_error(y_true=iris.petal_length, y_pred = iris_predict)
```
