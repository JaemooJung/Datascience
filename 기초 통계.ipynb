데이터가 비대칭적일때 <p>
log변환, root 변환, inverse 변환을 통해 값을 대칭적으로 만들어 줄 수 있음 <p>
왜냐? 대부분의 통계기법은 데이터가 대칭적이라는 가정 하에 만들어졌기때문...<p>
데이터를 보고 적절한 변수를 만들어 내는 것이 실력임...

변환 = Transformation

변환을 통해 똑같은 데이터를 다양한 모델에 적용할 수 있게 됨. 특히나 대칭이 아닌 데이터를 대칭을 가정한 모델에 적용할 수 있도록 바꿔주게 되는 것...

-> **Box-Cox Transformation**



```
# 2.4 새로운 변수 만들기(새로운 열 만들기)
# (1) 연산
# 새로운 변수명 : xyz_mean : (x + y + z) / 자유도 = 3
# data['new_variable'] = 연산 -> 데이터의 제일 오른쪽에 연산 결과열이 추가됨
diamonds['xyz_mean'] = (diamonds.x + diamonds.y + diamonds.z) / 3 
diamonds.head(3)
```

```
# (2) 변환(Transformation) : Box-Cox Transformaiton
# i.    log
# ii.   root
# iii.  inverse
```

```
# i. price -> price_log10
diamonds['price_1og10'] = np.log10(diamonds.price)
diamonds.head()
```

```
# ii. root
#price -> price_root
diamonds['price_root'] = np.sqrt(diamonds.price) #square root의 약자
diamonds.head()
```

```
# iii. inverst
# price -> price_inverse
diamonds['price_inverse'] = 1 / diamonds.price
diamonds.head()
```

```
# (3) 양적 자료 (수치형자료)를 질적자료(범주형 자료) 로 바꾸기
# carat -> carat_group
# 0~2 : 'light'
# 2~4 : 'middle'
# 4~6 : 'heavy'
diamonds['carat_grout'] = pd.cut(diamonds.carat,
                                 bins=[0,2,4,6], # -> 자르는 구간
                                 right = False, # -> 구간의 상한 = right, 하한 = left(False). 구간의 상한이 중복값을 가질 것인지, 상한이 중복값을 가질 것인지 정하는 파라미터 
                                 labels = ['light', 'middle', 'heavy']) 
diamonds.head()
```

구간(interval or mean)을 잘 자르는 법 -> 

1. 최소값과 최대값 찾기 ( np.max() , np.min() )
2. 범위(range)구하기 (max - min)
3. 구간의 갯수를 몇개로 할까 ? 근데 이게 사실 제일 힘든 사항이다... 
  그래서! <p>[ 1+3.3*log10(n) n = 데이터의 갯수 ] 라는 공식이 있음.
4. 구간의 너비(폭) = 범위/구간의 갯수
5. 첫번째 구간에 최소값이 오도록 조정, 마지막 구간에 최대값이 오도록 구간.

    상한 = 하한 + 구간

```
#2.7 데이터 합치기
# (1) Merge : inner join, outer join(full join), outer join(left join), outer join(right join)
# (2) Concat : axis = 0(row), axis = 1(column)
```

```
d1 = pd.DataFrame({'id':[1,2,4,5],
                   'age':[10,20,30,40],})
d1
```

```
d2 =  pd.DataFrame({'id':[1,3,4,6],
                   'bt':['a','b','ab','o'],})
d2
```

```
# merge - inner join -> primary key가 합치려고 하는 두 df 모두에 있는 것만 합침
# pd.merge(data1,data2, on= 'primary_key',how='inner')
d3 = pd.merge(d1,d2,on='id', how='inner')
d3
```

```
# merge - outer join(full join) -> 두 df를 합치는데 어느 한쪽에 값이 없는 쪽은 NaN으로 비워둠.
# pd.merge(data1,data2, on= 'primary_key',how='outer')
d4 = pd.merge(d1,d2,on='id',how='outer')
d4
d4.sort_values(by = 'id', ascending=True)
```

```
# merge - outer join(left join) -> 먼저 들어간 (left에 있는) df의 p_key가 있는 행만 합침
# pd.merge(data1, data2, on='primary_key', how='left')
d5 = pd.merge(d1,d2, on='id', how='left')
d5 
```

```
# merge - outer join(right join) -> 나중에 들어간 (right에 있는) df의 p_key가 있는 행만 합침
# pd.merge(data1, data2, on='primary_key', how='right')
d6 = pd.merge(d1,d2, on='id', how='right')
d6
```
