### 1. 일표본 t검정(One sample t-test)


```
# 모집단의 양적 자료가 정규분포를 따른다고 가정할 때 
```

```
# diamonds의 price
# 귀무가설 : 다이아몬드 가격의 평균은 4000달러이다
# 대립가설 : 다이아몬드 가격의 평균은 4000달러 이상이다.
```

```
# scipy.stats -> stats
# stats.ttest_1samp(data.variable, popmean = mu )

# 참고
# data.variable -> 검정 변수
# popmean -> 귀무 가설의 모평균(mu) 
```

```
stats.ttest_1samp(diamonds.price, popmean= 4000)
```

```
#파이썬에서 p값을 해석할 때 주의할 점 = 파이썬 p값은 t분포에서 양 끝 p값의 합을 구해주는 것임. 따라서 대립가설에 따라 p값의 범위를 다시 계산해봐야함.
```

```
#statistic = t값
# 평균 : diamonds.price.mean()
# 표준편차 =
price_mean = diamonds.price.mean()
price_std = diamonds.price.std()
price_n = diamonds.price.count()
price_t = (price_mean - 4000) / (price_std/np.sqrt(price_n))

print(price_mean)
print(price_std)
print(price_n)
print(price_t)
```

```
# 결론
# 유의확률(p-value)이 1.000 이므로 유의수준 0.05에서 '다이아몬드 가격의 평균은 4000달러이다' 라는 귀무가설을 채택한다.
# 다이아몬드 가격에는 통계적으로 유의미한 차이는 없는 것으로 나타났다.
```

```
# diamonds.carat
# 귀무가설 : 다이아몬드의 무게 평균은 1캐럿이다.
# 대립가설 : 다이아몬드의 무게 평균은 1캐럿보다 작다. 
```

```
diamonds.carat.describe()
```

```
stats.ttest_1samp(diamonds.carat, popmean=1)
```

```
# p-value = 0.000
# 따라서 유의수준 0.05에서 다이아몬드의 무게에 통계적으로 유의한 변화가 있는 것으로 나타났다.
# 다이아몬드의 무게가 통계적으로 유의하게 작아진 것으로 나타났다.
```

```
#복습
#diamonds의 x
# 귀무가설 : 다이아몬드의 x의 길이의 평균은 5이다.
# 대립가설 : 다이아몬드의 x 길이의 평균은 5가 아니다.
```

```
stats.ttest_1samp(diamonds.x, popmean = 5)
```

```
#결과
#p-value = 0.000
# 따라서 유의수준 0.05에서 다이아몬드의 x길이에는 통계적으로 유의한 변화가 있는 것으로 나타났다.
# 따라서 다이아몬드의 x길이의 평균은 5라는 귀무가설을 기각한다.
```

### 2. 윌콕슨 부호순위검정 (Wilcoxon's signed rank test)

```
# 모집단의 양적 자료가 정규분포를 따르지 않을 때
# 정규성 가정이 깨질 때
# 하지만 데이터의 고유 특성이 깨져버린다는 단점이 있음.
# 따라서 웬만한 상황에서는 t검정이 좋지만...
# 정규성이 깨진 데이터(이상치가 있는 데이터)는 윌콕슨 검정을 사용
# 보통은 데이터에서 이상치를 가공한 후 가공된 데이터를 t검정을 통해 검정하는 경우가 많음.
# 윌콕슨 검정보다 t검정보다 검정력(power)이 좋기 때문. 
```

```
# stats.wilcoxon(data.variable - mu , alternative = 'grater' or 'less' or 'two.sided')

# mu값을 빼주는 이유 = 윌콕스는 기본이 평균 = 0 이라는 가설 검정을 해주기 때문.
# 따라서 모평균(mu) 값을 빼주는 것. 
# alternative : 대립가설
# 'grater' : mu > 0 , 'less' : mu < 0 , 'two.sided': mu != 0
```

```
# 귀무가설 : 다이아몬드 가격의 평균은 4000달러이다
# 대립가설 : 다이아몬드 가격의 평균은 4000달러 이상이다
```

```
stats.wilcoxon(diamonds.price - 4000, alternative='greater')
```

```
# 결과 : 유의확률이 1.000 이므로 유의수준 0.05에서 귀무가설 채택
```

```
# diamonds의 carat
# 귀무가설 : 다이아몬드의 무게 평균은 1캐럿이다
# 대립가설 : 다이아몬드의 무게 평균은 1캐럿 이하이다.
stats.wilcoxon(diamonds.carat - 1, alternative='less')
```

```
# 결론
# 유의 확률이 0.000 이므로 유의수준 0.05에서
# 다이어몬드의 무게에 통계적으로 유의하게 작아진 것으로 나타났다.
```

```
# diamonds의 X
# 귀무가설 : 다이아몬드의 x길이 평균은 5이다.
# 대립가설 : 다이아몬드의 x길이 평균은 5가 아니다.
```

```
stats.wilcoxon(diamonds.x - 5, alternative='two-sided')
```

```
# 결론 유의 확률 0.000으로 유의수준 0.05에서 다이아몬드의 x길이에는 통계적으로 유의미한 변화가 있는 것으로 나타났다.
# 귀무가설을 기각하고 대립가설을 채택한다
```

### 정규성 검정(Normality test)

```
# 귀무가설 : 모집단의 양적 자료는 정규분포를 따른다
# 대립가설 : 모집단의 양적 자료는 정규분포를 따르지 않는다.
```

```
# 1. Shapiro-Wilk Normality test
# 데이터의 갯수가 미만일 때 사용. 5000개 미만일 때 가장 강력하기 때문.
# stats.shapiro(data.variable)
stats.shapiro(diamonds.price)
```

```
# 이 경우 p-value가 0.000이므로 
```

```
# 2. Anderson-Darling Normality Test
# 데이터의 갯수가 5000개 이상일 때 사용
# stats.anderson(data.variable, dist = 'norm')
# 참고
# dist = distribution = 분포
# 'norm' = normal distribution = 정규분포
# 참고 2
# 표본에 있는 데이터의 갯수 (표본집단의 크기) = sample size

stats.anderson(diamonds.price, dist='norm')
```

```
# Significance_level = 유의수준
# Critical_values = 임계값
# 유의 수준별 임계값을 보여줌
# 유의수준 0.05에서, 0.787보다 높은 값을 만나면 대립가설 채택, 아닐 시 귀무가설 채택.
# 귀무 = 정규분포다 / 대립 = 정규분포가 아니다
# 따라서 다이아몬드 가격은 정규분포를 따르지 않는다.
# 유의수준 0.05에서의 임계값이 0.787이고
# 표본에서 검정통계량이 3474.016으로 임계값 0.787보다 크므로
# price는 정규분포를 따르지 않는다. 즉 정규성 가정이 깨짐.
```

### 일표본 검정의 단계


```
# 1단계 : 정규성 검정
# stats.shapiro(data.variable) 데이터의 크기가 5000개 이하
# stats.anderson(data.variable, dist='norm') 5000개 이상
# 귀무가설 : 정규분포를 따른다.
# 대립가설 : 정규분포를 따르지 않는다.
```

```
# 2단계 : 일표본 t검정(One sample t-test)
# 조건 : 1단계의 결론이 귀무가설, 즉 정규성 가정을 만족
# stats.ttest_1samp(data.variable, popmean= )
```

```
# 2단계 : 윌콕슨 부호순위 검정(Wilcoxon's signed rank test)
# 조건 : 1단계의 결론이 대립가설, 즉 정규성 가정이 깨짐
# stats.wilcoxon(data.variable - mu, alternative = '')
```

```
# 연습 - diamonds의 price
# 귀무가설 : 다이아몬드 가격의 평균은 4000달러이다.
# 대립가설 : 다이아몬드 가격의 평균은 4000달러보다 높다.
```

```
stats.anderson(diamonds.price)
```

```
# 다이아몬드의 가격은 정규분포를 따르지 않음.
# 검정 통계량 = 3474.016 이므로 유의수준 0.05의 임계값 0.787보다 크므로 정규성 가정은 깨진다
```

```
# 윌콕슨 검정 사용
stats.wilcoxon(diamonds.price - 4000, alternative = 'greater')
```

```
# 결론 : 유의확률이 1.000이므로 유의수준 0.05에서 다이아몬드 가격은 통계적으로 유의하게 변하지 않았다.
```
